there are many columns, now is time to check quality of data, ie any null 

```python
# https://stackoverflow.com/questions/53283142/understand-df-isnull-mean-in-python
# why is isnull.mean 

features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]

for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(),4), '  missing values')
```

```python
a = dataset['LotFrontage'].isnull().mean()
print(a)
```

find relationship between missing values and saleprice

```python
for feature in features_with_na:
    data = dataset.copy()
    
#     put 1 if null , 0 if not
    data[feature] = np.where(data[feature].isnull(),1,0)
#     calculate sale price where info is missin or present
# https://www.geeksforgeeks.org/python-pandas-dataframe-groupby/
    data.groupby(feature)['SalePrice'].median().plot.bar(color=colors)
    plt.title(feature)
    plt.show()
    
```

take one step down, check all year with price

```python
for feature in year_feature:
    if feature != 'YrSold':
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

continuous variable

```python
continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
```

```python
for feature in continuous_feature:
    data=dataset.copy()
#     Calculate the number of bins by taking the square root of the number of data points and round up.
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
# The log transformation can be used to make highly skewed distributions less skewed
for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

outliers

```python
for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
